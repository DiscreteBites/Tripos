```python
## example sheet 1 question 11 tests
def T(m, n):
    if m == 1:
        return n + 1
    
    elif m > 1 and n == 1:
        return T(m-1, 2)

    elif m > 1 and n > 1:
        return T(m-1, T(m, n-1))
```

```python
def test(m, n):
    if m == 3:
        return 2*n + 2

    if m == 4:
        return 2**(n + 2) - 2
    
    if m == 5:
        a = 16
        for _ in range(n-1):
            a = 2**(a)
        return a -2
    
```

```python
case = 4

for i in range(1, 10):
    a = T(case, i)
    b = test(case, i)
    print(f'{i}: {a} {b} {a == b}')
```

```python
N = 100

# Helper functions to regenerate the sums/products data structures every round.
# Not optimal but hey this is for fun.
def generate_sums(candidate_pairs):
    sums = {}
    for tuple_pair in candidate_pairs:
        (i, j) = tuple_pair
        sum = i + j
        if sum not in sums:
            sums[sum] = [tuple_pair]
        else:
            sums[sum].append(tuple_pair)
    return sums

def generate_products(candidate_pairs):
    products = {}
    for tuple_pair in candidate_pairs:
        (i, j) = tuple_pair
        product = i * j
        if product not in products:
            products[product] = [tuple_pair]
        else:
            products[product].append(tuple_pair)
    return products
```

```python
candidate_pairs = set()
for i in range(2, N):
    for j in range(i +1, N):
        candidate_pairs.add((i, j))

sums = generate_sums(candidate_pairs)
products = generate_products(candidate_pairs)
```

```python
# Sophie can tell that paul dosent know the numbers
sophie_candidates = {}

for sum, candidate_pairs in sums.items():
    incl = True

    if len(candidate_pairs) == 1:
        incl = False
    
    for c in candidate_pairs:
        if len(products[c[0] * c[1]]) == 1:
            incl = False
            break
    
    if incl:
        sophie_candidates[sum] = candidate_pairs
```

```python
# batched coupon problem 
#
# https://math.stackexchange.com/questions/3278200/iteratively-replacing-3-chocolates-in-a-box-of-10/3278285#3278285
#
import scipy.special
from math import factorial
from fractions import Fraction

def binom(n,k): # binomial coefficient n choose k
  if n < k :
    return 0
  
  return scipy.special.comb(n, k, exact=True)
  return int(factorial(n)/factorial(k)/factorial(n-k))

def e(n,m): # expected number of days to replace all n chocolates if replaced m at a time
  return binom(n,m)*sum(Fraction(pow(-1,s+1)*binom(n,s),binom(n,m)-binom(n-s,m)) for s in range(1,n+1))

e = e(357,100)
print(float(e))
```
